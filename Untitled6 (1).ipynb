### Download and Exploar data

before we go on with the project we need to understand what is data we need for the project. as I wrote in the introduction the purpose of the project is to give an overview of the neighborhoods in New York City. to do so we will use the Foursquare API to get the coordinates of the neighborhoods in New York City. but first, we need to get the latitude and longitude of New York City 

```python
address = 'New York, NY'

geolocator = Nominatim(user_agent="foursquare_agent")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print(latitude, longitude)
```

After we get the coordination of New York City. We use the foursquare API to get all coordinates of the Italian restaurant in the neighborhoods.

```python
import requests

results = requests.get('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DS0701EN-SkillsNetwork/labs/newyork_data.json')


j = results.json()
j

#df=pd.DataFrame.from_dict(j)
```

since we have the data as a json we nead to convert it into pandas dataframe

```python
neighborhoods_data = j['features']

column_names = ['Borough', 'Neighborhood', 'Latitude', 'Longitude'] 
neighborhoods = pd.DataFrame(columns=column_names)

for data in neighborhoods_data:
    borough = neighborhood_name = data['properties']['borough'] 
    neighborhood_name = data['properties']['name']
        
    neighborhood_latlon = data['geometry']['coordinates']
    neighborhood_lat = neighborhood_latlon[1]
    neighborhood_lon = neighborhood_latlon[0]
    
    neighborhoods = neighborhoods.append({'Borough': borough,
                                          'Neighborhood': neighborhood_name,
                                          'Latitude': neighborhood_lat,
                                          'Longitude': neighborhood_lon}, ignore_index=True)
```

## Results

since we have the dataframe with the cordinats we need to show the neighborhood in the New York City map. to do so we use the  folium.Map() using the coordination we get for New York City.

```python
NYC_map = folium.Map(location=[latitude, longitude], zoom_start=10.5, )

NYC_map
```

after giting the map of New York City we use folium agan to marck the neighborhood.

```python
NYC_neighborhood = folium.map.FeatureGroup()

# loop through the 100 crimes and add each to the incidents feature group
for lat, lng, label, in zip(neighborhoods.Latitude, neighborhoods.Longitude, neighborhoods.Neighborhood):
    NYC_neighborhood.add_child(
        folium.features.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )    

NYC_map.add_child(NYC_neighborhood)
```
